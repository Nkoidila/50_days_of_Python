```python

import requests
```

```python
# Query an external API to find geographical information about a given IP address
# Tasks: 1.Prompt the user for an IP address 2.Send a request to a geolocation API 3.Display City,Region and Country


# Prompt user for IP address
ip_address = input("Hello, please enter a valid IP address: ")

try:
    # Send request to API
    url = f"http://ip-api.com/json/{ip_address}"
    response = requests.get(url)

    data = response.json()

    # Check if request was successful
    if data["status"] == "success":
        print("\n📍 Location:")
        print("City:", data["city"])
        print("Region:", data["regionName"])
        print("Country:", data["country"])
    else:
        print("❌ Invalid IP address or request failed")

except Exception as e:
    print("Error:", e)
```
