```python
# Create a dictionary where student names are keys and grade as values
# Allows users to: 1.Add a new student 2.Update grade 3.Retrieve grade 4.Display all students

#Storing initial students data
students={"Saka":85,
          "Raya":92,
          "Martin":78,
          "Martinelli":84,
          "Zubimendi":72,
          "White":45,}

#Function for adding new student
def add_student(name,grade):
    students[name]=grade

#Function for updating student grade
def update_grade(name,new_grade):
    if name in students:
        students[name]=new_grade
    else:
        print("Student not found")

#Retrieving grade
def retrieve_grade(name):
    if name in students:
        return students[name]
    else:
        print("Student not found")

#Displaying all students
def display_students():
    for name,grade in students.items():
        print(f"{name}:{grade}")
```

```python
#adding a new student
add_student("Timber",88)
#print(students)

#updating a students grade
update_grade("Raya",97)
#print(students)

#retrieving a students grade
print(retrieve_grade("Raya"))

#displaying all students
display_students()
```
