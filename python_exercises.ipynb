```python
#37. Counting word frequency in a sentence
sentence="For Arsenal to win the league, Arsenal must win the seven of the remaining league games so as not to be considered the Arsenal bootlejobs"
from collections import Counter
words = sentence.lower().split()

word_count = Counter(word.strip(",.") for word in words)

print(word_count)
```
